```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.uniform(-5,5, size=(10,2))
data
```

```python
plt.figure(dpi=250)
plt.scatter(data[:,0], data[:,1]);
plt.title("Books in a 'Genre' Space")
plt.xlabel("Horror")
plt.ylabel("Sci-Fi");
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15,15), dpi=250)
ax = fig.gca(projection='3d')
data = np.random.uniform(-5,5, size=(30,3))
ax.scatter(data[:,0], data[:,1], data[:,2], s=60, alpha=0.8);
ax.set_title("Books in a 'Genre' Space")
ax.set_xlabel("Horror", fontsize=20, labelpad=20)
ax.set_ylabel("Sci-Fi", fontsize=20, labelpad=20)
ax.set_zlabel("Romance", fontsize=20, labelpad=20);
```
